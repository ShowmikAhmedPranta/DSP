```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
srate=256
t=np.arange(0.,1.,1/srate)
x1=3*np.sin(2*np.pi*10*t)

x2=3*np.cos(2*np.pi*10*t)

x3=x1+x2

DC=7

x4=x1+x2+DC
```

```python
plt.figure(figsize=(24,14))
#plt.figure()
plt.suptitle("Constructing a wave with sine,cosine and DC waves",fontsize=40)

plt.subplot(2,2,1)
plt.plot(t,x1,'-r',linewidth=3)
plt.title('Sinewave of amplitute 3 and frequency 10hz',fontsize=25)
plt.subplot(2,2,2)
plt.plot(t,x2,'-g',linewidth=3)
plt.title('coSinewave of amplitute 3 and frequency 10hz',fontsize=25)
plt.xlabel('time in sec')
plt.ylabel('Amplitude')

plt.subplot(2,2,3)
plt.plot(t,x3,'-b',linewidth=3)
plt.title('sum of sine , cosine',fontsize=25)

plt.subplot(2,2,4)
plt.plot(t,x4,'-y',linewidth=3)
plt.title('Combined sine wave,cosine,DC',fontsize=25)
plt.xlabel('time in sec',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
srate=256
t=np.arange(0.,1.,1/srate)

x1=3*np.sin(2*np.pi*10*t)

x2=3*np.cos(2*np.pi*10*t)

x3=x1+x2

DC=7

x4=x1+x2+DC

#Plotting the signal and its fourier transform


plt.figure(figsize=(24,14))
#plt.figure()
plt.suptitle("Signal and its fourier transform",fontsize=40)

plt.subplot(2,2,1)
plt.plot(t,x4,linewidth=3)
plt.title('Sinewave of amplitute 3 and frequency 5hz',fontsize=25)
plt.xlabel('time in sec')
plt.ylabel('Amplitude')

plt.subplot(2,2,2)
amps=np.abs(x4)/len(t)
plt.stem(amps,t)
plt.title('Fourier transform of a signal',fontsize=25)
plt.xlabel('indices')
plt.ylabel('Amplitude')

plt.show()
```
