```python
t  = np.linspace(0,10,150) # time 
f = .2;                    # frequency 
A = 23;                    # amplitude 
theta = np.pi/3;           # phase 
```

```python
Compx_wave = A * np.exp(-5j* (2*np.pi * f * t + theta) );
```

```python
plt.figure()
plt.plot(t,Compx_wave,'g-',linewidth = 3)
plt.xlabel('time',fontsize = 15)
plt.ylabel('amplitude', fontsize = 15)
plt.title('Complex wave',fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
plt.figure()
plt.plot(t,np.real(Compx_wave),'b-',linewidth = 3, label='real part')
plt.xlim([-0.1,5])
plt.xlabel('time',fontsize = 15)
plt.ylabel('amplitude', fontsize = 15)
plt.title('Real sine wave',fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
plt.figure()
plt.plot(t,np.imag(Compx_wave),'m-',linewidth = 3, label='imaginary part')
plt.xlim([-0.1,5])
plt.xlabel('time',fontsize = 15)
plt.ylabel('amplitude', fontsize = 15)
plt.title('Imaginary sine wave',fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
plt.figure()

plt.plot(t,np.imag(expr),'b-', label='imaginary part')
plt.plot(t,np.real(expr),'r-', label='real part')

plt.xlim([-0.1,30])
plt.xlabel('time',fontsize = 15)
plt.ylabel('amplitude', fontsize = 15)
plt.title('Real and imaginary sine wave',fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```
