```python
import numpy as np, matplotlib.pyplot as plt, matplotlib.style as st
```

```python
#Define analog signal
t=np.arange(0,11)
x=np.sin(t)
```

```python
t
```

```python
x
```

```python
#plotting analog signal
plt.plot(t,x)
plt.xlabel("time,t")
plt.ylabel("amplitude,x")
plt.title("Analog signal")
plt.legend(['x=sin(t)'])
plt.show()
```

```python
#Sampling and plotting sampled signal
plt.plot(t,x)
plt.stem(t,x)
plt.xlabel("time,t")
plt.ylabel("amplitude,x")
plt.title("Sampled signal")
plt.xticks(t)
plt.yticks(x)
plt.legend(['x=sin(t)'])
plt.show()
```

```python
#Quantization and plotting quantized signal
Qx=np.around(x,1)
plt.stem(t,Qx)
plt.xticks(t)
plt.yticks(Qx)
plt.xlabel("time,t")
plt.ylabel("amplitude,x")
plt.legend(['x=sin(t)'])
plt.title("Quantized signal")
for i in Qx:
    plt.axhline(i)
plt.show()
```

```python
#print t vs Qx
print("t --->Qx")
for i,j in zip(t,Qx):
    print(i,"--->",j)
```

```python
#define analog signal
import math
t1=np.arange(-10,10)
x1=2*np.sin(t1/2)+2
```

```python
x1,t1
```

```python
#plotting analog signal
plt.plot(t1,x1)
plt.xlabel("time,t")
plt.ylabel("amplitude,x")
plt.title("Analog signal")
plt.legend(['2sin(t/2)+2'])
plt.show()
```

```python
#Sampling and plotting sampled signal
plt.figure(figsize=(10,15))
plt.plot(t1,x1)
plt.stem(t1,x1)
plt.xlabel("time,t")
plt.ylabel("amplitude,x")
plt.xticks(t1)
plt.yticks(x1)
plt.title("Sampled Signal")
plt.legend(['2sin(t/2)+2'])
plt.show()
```

```python
#Quantization and plotting quantized signal
Qx1=np.around(x1,1)
plt.figure(figsize=(7,15))
plt.stem(t1,Qx1)
plt.xticks(t1)
plt.yticks(Qx1)
plt.legend(['2sin(t/2)+2'])
for i in Qx1:
    plt.axhline(i)
plt.title("Quantized Signal")
plt.show()
```

```python
#print t vs Qx
print("t --->Qx")
for i,j in zip(t1,Qx1):
    print(i,"--->",j)
```
